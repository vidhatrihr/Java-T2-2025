## Lect-1

* Programming languages are *human readable* and used for computation, this high level language is converted to low level language so that computers can understand.
* A programming language is a medium for communication between humans and machines.
  
/ *Compiler:*
- Translates the whole source code into machine code before running the program.
- Fast execution, but slower to compile.
- Example: C, C++.

/ *Interpreter:*
- Translates and executes code line by line at runtime.
- Slower execution, but no need to compile first.
- Example: Python, JavaScript.

* *Declarative* (what)
* *Imperative* (how)

## Lect-4

* Memory Allocation at Runtime 
  - When a program runs, memory is divided into regions
    - Code Segment (Text Segment)
    - Static/Global Segment (Data Segment)
    - Heap Segment (Dynamic Memory)
    - Stack Segment (Function Call Stack)
    -  Register (CPU Level, Not RAM)

* Data Abstraction means showing only essential data and hiding unnecessary details from the user.

* *Polymorphism:* The ability of the same function, method, or operator to behave differently based on the object or data it's acting on.

* *Modularity*: breaking a program into smaller, independent, self-contained units (modules)
